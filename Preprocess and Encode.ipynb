```python
# Import Required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Import data
data = pd.read_csv("./data.csv")
data.head()
```

```python
# Get information about null values and data types of the columns
data.info()
```

```python
# Encode the columns using LabelEncoder
for columns in data.columns:
    labelencoder = LabelEncoder()
    if columns not in ['Name'] and data[columns].dtypes == 'O':
        data[columns] = labelencoder.fit_transform(data[columns])
        print(labelencoder.classes_)
```

```python
# Save it to encoded_data.csv
data.to_csv("encoded_data.csv")
```
