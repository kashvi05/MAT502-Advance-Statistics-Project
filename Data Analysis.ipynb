```python
# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Import data
data = pd.read_csv("./encoded_data.csv")
data.head()
```

```python
data.info()
```

```python
# Plot Maximum willingness to pay before and after covid for each respondent
from matplotlib.lines import Line2D
    
g = sns.lineplot(data=data.MWP_BC, color="g")
sns.lineplot(data=data.MWP_AC, color="b", ax=g.axes.twinx())
g.legend(handles=[Line2D([], [], marker='_', color="g", label='MWP_BC'), Line2D([], [], marker='_', color="b", label='MWP_AC')])
plt.title("Maximum Willingness to pay before and after covid")
```

```python
# Plot Quantity of bottles used before and after covid for each and every respondent

from matplotlib.lines import Line2D
g = sns.lineplot(data=data.MQB_BC, color="g")
sns.lineplot(data=data.MQB_AC, color="b", ax=g.axes.twinx())
g.legend(handles=[Line2D([], [], marker='_', color="g", label='MQB_BC'), Line2D([], [], marker='_', color="b", label='MQB_AC')])
plt.title("Quantity of bottles used before and after covid")
```

```python
# Convert gender to numeric data type
data["Gender"]= pd.to_numeric(data["Gender"], downcast="float")
```

```python
# Count the number of values and store it in different dataframe
gender = data['Gender'].value_counts()
```

```python
# Construct a pie chart for gender

mylabels_gender =["Males","Females"]
plt.pie(gender, labels=mylabels_gender,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
brand = data['Brand'].value_counts()
```

```python
# Construct a pie chart for Brands of sanitizers

mylabels_brand = ["Dettol","Lifebuoy","Himalaya","Savlon","Any","Godrej","Cipla", "Coax"]
plt.pie(brand, labels=mylabels_brand,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
type = data['Type'].value_counts()
type
```

```python
# Construct a pie chart for types of sanitizers
mylabels_type = ["Alcohol based","Gel based","Any"]
plt.pie(type, labels=mylabels_type, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
age = data['Age'].value_counts()
age
```

```python
# Construct pie chart for different age distributions

ylabels_age = ["16-20","31 or above","below 16","21-25"]
plt.pie(age, labels=ylabels_age,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
# Plot kde plots or gaussian distributions for willingness to pay before and after pandemic
columns = ['MWP_BC', 'MWP_AC']

for column in columns:
    plt.figure()
    sns.kdeplot(data=data, x=column)
    plt.show()
```
