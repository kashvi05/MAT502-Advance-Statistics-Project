```python
# Import required libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy 
import pandas as pd
import scipy.stats
```

```python
# Import data

data = pd.read_csv("./encoded_data.csv")
data.drop("Unnamed: 0", axis = 1, inplace=True)
data.head()
```

```python
# See statistics of all columns

data.describe()
```

```python
# Calculate standard error of maximum willingness to pay before covid

SE=np.std(data.MWP_BC)/np.sqrt(np.size(data.MWP_BC))
```

z_value = (observed - expected) / SE

```python
# Define a function to calculate p-value based on z-statistics

def p_value():
    z=(np.mean(data.MWP_AC)-np.mean(data.MWP_BC))/SE
    print("Z value= ",z)
    p=scipy.stats.norm.sf(abs(z))
    print("p_value= ",p)
```

```python
# Calculate p-value for conclusions

p_value()
```

Z-value tells us how many std errors is the observed value from the expected value. The z-value is very high which tells us that the observed value is very far from the expected value and as a result we get a low p-value.Hence, we reject the null hypothesis that says that the variation is NOT due to chance error.
